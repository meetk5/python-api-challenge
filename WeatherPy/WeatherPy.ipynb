```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from datetime import date

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).

```python
city = []
lat = []
long = []
maxtemp = []
humidity = []
cloudiness = []
windspeed = []
country = []
date = []
units = "imperial"

for i in range(len(cities)):
    query_url = f"http://api.openweathermap.org/data/2.5/weather?q={cities[i]}&units={units}&APPID={weather_api_key}"
    print(query_url)
    response = requests.get(query_url).json()
    try:
    #currentweather.append(response.json())
        city.append (response["name"])
        lat.append (response["coord"]["lat"])
        long.append(response["coord"]["lon"])
        maxtemp.append(response["main"]["temp_max"])
        humidity.append(response["main"]["humidity"])
        cloudiness.append(response["clouds"]["all"])
        windspeed.append(response["wind"]["speed"])
        country.append(response["sys"]["country"])
        #city_id.append(response["id"])
        date.append(response["dt"])
        print(f"Processing Record {i+1} of Set 1 | {cities[i]}")
        i = i + 1
   # if i == 60 or i == 120 or i ==180 or i == 240 or i == 300 or i == 360 or i == 420 or i == 480:
        time.sleep(2)
 
    except KeyError:
        print("City not found!! Skipping...")
        pass
    if i == 550: 
        break
        
print(len(city))
```

```python
weatherdf = pd.DataFrame({"City": city,"Latitude":lat,"Longitude":long,"Max Temp": maxtemp,"Humidity": humidity,
                         "Cloudiness": cloudiness,"Wind Speed": windspeed,"Country": country,"Date": date})
weatherdf.head()
```

## Linear Regression

```python
nh_df = weatherdf.loc[weatherdf["Latitude"]>0]
nh_df.head()
```

```python
#Linear Regression function for Northern Hemisphere
def linregNorth(yvalue):
    x_values = nh_df["Latitude"]
    y_values = nh_df[yvalue]

    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values,y_values)
    print("The r-value is ",rvalue)

    regress_values = x_values* slope + intercept

    plt.scatter(x_values,y_values)
    plt.title(f"Northern Hemisphere - {yvalue} vs. Latitude Linear Regression")

    line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
    plt.plot(x_values, regress_values, color = "red")

    plt.xlabel("Latitude")
    plt.ylabel(yvalue)
    plt.annotate(line_eq,(5,40), color = "red",fontsize = 15)

    plt.savefig(f"../output_data/ScatterPlt-{yvalue} vs LAT-NorthHemi")
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
linregNorth("Max Temp")
```

******The rvalue indicates that Latitude and Temperature are negatively correlated. i.e. the lower the latitide, the higher the temperature.
This clearly proves the hypothesis that the temperature is the highest at the equator (where latitude is 0). The temperature keeps getting lower as we move to the higher latitude i.e. near the North pole.

```python
sh_df = weatherdf.loc[weatherdf["Latitude"]<0]
sh_df.head()
```

**Linear Regression function for Southern Hemisphere

```python
#Linear Regression function for Southern Hemisphere
def linregSouth(y_axis):
    x_values = sh_df["Latitude"]
    y_values = sh_df[y_axis]

    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values,y_values)
    print("The r-value is ",rvalue)

    regress_values = x_values* slope + intercept

    plt.scatter(x_values,y_values)
    plt.title(f"Southern Hemisphere - {y_axis} vs. Latitude Linear Regression")

    line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
    plt.plot(x_values, regress_values, color = "red")
    
    plt.xlabel("Latitude")
    plt.ylabel(y_axis)
    plt.annotate(line_eq,(-50,90), color = "red",fontsize = 15)

    plt.savefig(f"../output_data/ScatterPlt-{y_axis} vs LAT-SouthHemi")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
linregSouth("Max Temp")
```

****The r-value and the line plot above clearly prove that the temperature and the latitude are correlated.
As we go below the equator (where latitude is 0), the temperature keeps falling. As the latitude is lower than 0 in the Southern Hemisphere, the temperature keeps keeps getting lower. 
This strengthens our hypothesis that it's the hottest at the equator and coldest at the poles.

**The Humidity and Latitude are slightly negatively correlated. This proves that it gets drier as we move towards the poles. However, latitude does not have a significant impact on the humidity in Northern Hemisphere.

####  Northern Hemisphere - Wind Speed (mph) vs. Latitude Linear Regression

```python
x_values = nh_df["Latitude"]
y_values = nh_df["Wind Speed"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values,y_values)
print("The r-value is ",rvalue)
    
regress_values = x_values* slope + intercept

plt.scatter(x_values,y_values)
plt.title("Northern Hemisphere - Wind Speed vs. Latitude Linear Regression")

line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
plt.plot(x_values, regress_values, color = "red")

plt.xlabel("Latitude")
plt.ylabel("Wind Speed")
plt.annotate(line_eq,(5,25), color = "red",fontsize = 15)

plt.savefig("../output_data/ScatterPlt-Wind Speed vs LAT-NorthHemi")
```

**The low r-value and the Linear Regression plot indicate that an insignificant correlation between the Latitude and Wind Speed in Northern Hemisphere. There's a mild negative correlation though, which means that as we move away from Equator, Wind speed increases.

####  Southern Hemisphere - Wind Speed (mph) vs. Latitude Linear Regression

```python
x_values = sh_df["Latitude"]
y_values = sh_df["Wind Speed"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values,y_values)
print("The r-value is ",rvalue)

regress_values = x_values* slope + intercept

plt.scatter(x_values,y_values)
plt.title("Southern Hemisphere - Wind Speed vs. Latitude Linear Regression")

line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
plt.plot(x_values, regress_values, color = "red")

plt.xlabel("Latitude")
plt.ylabel("Wind Speed")
plt.annotate(line_eq,(-50,20), color = "red",fontsize = 15)

plt.savefig("../output_data/ScatterPlt-Wind Speed vs LAT-SouthHemi")
```

**Just like the Northern Hemisphere, there is not a strong correlation between Wind Speed and Latitude even in the Southern Hemisphere. The low r-value and the Linear Regression plot above prove that. 
In a nutshell, there is only a strong correlation between the Temperature and the Latitude. As we go away from the Equator, the temperature decreases (as clearly visible from the scatter plots and the r-values in both hemispheres).
The rest of the factors like Wind Speed, Cloudiness and Humidity are not strongly correlated with the Latitude.
